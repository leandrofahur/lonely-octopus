### #2 step: Preprocessing
#### Exploring the statistics of the variables

```python
# raw_data.describe()
raw_data.describe(include='all')
```

#### Determining the variables of interest

```python
# data = raw_data.drop(['Model'], axis=1)
data = raw_data.drop(columns=['Model'])
data
```

#### Dealing with missing values

```python
# data.isnull()
data.isnull().sum()
```

```python
# drop the rows with missing values:
data_no_mv = data.dropna(axis=0)
data_no_mv.describe(include='all')
```

#### Exploring the Probability Distribution Functions (PDFs)

```python
sns.displot(data_no_mv['Price'])
```

#### Dealing with outliers

```python
q = data_no_mv['Price'].quantile(0.99)
data_1 = data_no_mv[data_no_mv['Price'] < q]
data_1.describe(include='all')
```

```python
sns.displot(data_1['Price'])
```

```python
sns.displot(data_no_mv['Mileage'])
```

```python
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage'] < q]
data_2.describe(include='all')
```

```python
sns.displot(data_2['Mileage'])
```

```python
sns.displot(data_no_mv['EngineV'])
```

```python
# some people put 99.99 in tables to symbolize an error or a no data:
data_3 = data_2[data_2['EngineV'] < 6.5]
```

```python
sns.displot(data_3['EngineV'])
```

```python
sns.displot(data_no_mv['Year'])
```

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year'] > q]
data_4.describe(include='all')
```

```python
sns.displot(data_4['Year'])
```

```python
data_cleaned = data_4.reset_index(drop=True)
```
