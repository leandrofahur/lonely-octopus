## <ins>Decomposition of variability: Unraveling Data Dynamics<ins>
Decomposition of variability allows us to dissect how much of the outcome's variation our model explains versus what remains as error.

### Example Code below:<br>
* Explanation: The Total Sum of Squares (TSS) represents the total variance in customer_satisfaction.
> The Regression Sum of Squares (RSS) quantifies how much of that variance is explained by the model.
<br>
> The Error Sum of Squares (ESS) is the variance the model fails to explain.

```python
import numpy as np
import statsmodels.api as sm

# Dataset: Product Quality Rating (independent) vs Customer Satisfaction (dependent)
quality_rating = [1, 2, 3, 4, 5]  # Independent variable
customer_satisfaction = [60, 65, 70, 75, 80]  # Dependent variable

X = sm.add_constant(quality_rating)  # Adding a constant for the intercept
y = customer_satisfaction

model = sm.OLS(y, X).fit()

# Decomposing variability
predictions = model.predict(X)
TSS = np.sum((y - np.mean(y))**2)
RSS = np.sum((predictions - np.mean(y))**2)
ESS = np.sum((y - predictions)**2)

print(f"Total Sum of Squares (TSS): {TSS}")
print(f"Regression Sum of Squares (RSS): {RSS}")
print(f"Error Sum of Squares (ESS): {ESS}")
```

## <ins>Multiple linear regression Theory<ins>
Multiple Linear Regression extends simple linear regression to incorporate multiple independent variables. The formula for a Multiple Linear Regression model is:

y = b0 + b1*x1 + b2*x2 + ... + bn*xn + e

where:<br>
* y is the dependent variable
* x1, x2,...xn are the independent variables,
* b0, b1,..., bn are the coefficients
* and e represents the error term

The goal of Multiple Linear Regression is to determine the coefficients that minimize the difference between the actual and predicted values of the dependent variable.

### Example Code below:<br>
* Explanation: The code exemplifies how to conduct multiple linear regression in Python using the sklearn library. It imports necessary libraries, prepares a dataset, splits it into training and testing sets, trains the model, and prints the model's coefficients and intercept.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Sample dataset with multiple independent variables
data = {'Feature1': [1, 2, 3, 4, 5], 'Feature2': [2, 4, 5, 7, 9], 'Scores': [2, 4, 5, 7, 9]}
df = pd.DataFrame(data)

# Splitting dataset
X = df[['Feature1', 'Feature2']]  # Independent variables
y = df['Scores']  # Dependent variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Multiple Linear Regression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Printing coefficients
print("Coefficients:", regressor.coef_)
print("Intercept:", regressor.intercept_)
```

## <ins>OLS assumptions<ins>
Standard deviation is a measure of the amount of variation or dispersion in a set of values. The coefficient of variation compares the standard deviation to the mean, providing a normalized measure of dispersion.

### Example Code below:<br>
* Explanation: These examples demonstrate how to calculate variance, standard deviation, and coefficient of variation for different datasets.

```python
import pandas as pd

# Sample data for individual weights in kilograms
weights = [65, 70, 68, 72, 75, 73, 71, 69, 67, 66, 74, 68, 70, 72, 71, 73, 69, 67, 75, 72]

# Creating DataFrame
df_weights = pd.DataFrame({'Weight(kg)': weights})

# Calculating standard deviation
std_deviation = df_weights['Weight(kg)'].std()
std_deviation1 = '{:.2f}'.format(std_deviation)

# Calculating mean weight
mean_weight = df_weights['Weight(kg)'].mean()

# Calculating coefficient of variation
cv = std_deviation / mean_weight
cv = '{:.1f}'.format(cv)

print(f"Standard Deviation: {std_deviation1}, Coefficient of Variation: {cv}, Mean: {mean_weight}")
```

## <ins>A2: No Endogeneity: Predictor-Error Independence<ins>
Detecting endogeneity involves scrutinizing the independence between predictors and the error term. While statistical tests for endogeneity (like the Hausman test) are complex, initial visual inspections can provide clues. If predictors are correlated with errors, it may manifest as patterns in plots of residuals against predictors, suggesting that crucial variables might be omitted or incorrectly measured.

### Example Code below:<br>
* Explanation: The provided code aims to assess the potential endogeneity in a regression model by conducting the Durbin-Wu-Hausman test.
> It begins by fitting an Ordinary Least Squares (OLS) regression model using synthetic data representing advertising expenditure, pricing strategy, customer satisfaction, and total sales.
<br>
> The endogeneity test is then performed using the Durbin-Wu-Hausman test, which compares the OLS estimator with a more efficient but potentially biased estimator.<br>
<br>
> The p-value of the test indicates whether the exogenous variables are correlated with the error term, providing insights into potential endogeneity issues in the regression model.

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_white

# Generate synthetic dataset
advertising_expenditure = [105, 98, 110, 102, 112]
pricing_strategy = [48, 52, 50, 55, 45]
customer_satisfaction = [80, 70, 85, 75, 90]
total_sales = [160, 155, 170, 165, 175]

# Fit OLS model
exog = sm.add_constant(list(zip(advertising_expenditure, pricing_strategy)))
model = sm.OLS(total_sales, exog)
results = model.fit()

# Check for endogeneity using Durbin-Wu-Hausman test
hausman_test = het_white(results.resid, exog)

print("Hausman test p-value:", hausman_test[1])
```

## <ins>A5: No multicollinearity<ins>
Understand the concept of multicollinearity among predictor variables and its adverse effects on regression analysis.

### Example Code below:<br>
* Explanation: Learn methods to detect multicollinearity using techniques like variance inflation factor (VIF) calculations.


* The Variance Inflation Factor (VIF) measures the extent of multicollinearity in a regression model.
> Multicollinearity occurs when two or more predictor variables in the model are highly correlated with each other.
<br>
> The VIF quantifies how much the variance of the estimated regression coefficients is inflated due to multicollinearity.The greater the value, the higher the multicollinearity.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Dataset for predictor variables
independent_variable_1 = [10, 12, 15, 18, 20, 22, 25, 28, 30, 32]  # Independent variable 1
independent_variable_2 = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65]  # Independent variable 2

# Create a matrix of predictor variables
exog = np.column_stack((independent_variable_1, independent_variable_2))

# Calculate VIF for predictor variables
vif = [variance_inflation_factor(exog, i) for i in range(exog.shape[1])]

print("VIF for predictor variables:", vif)
```

## <ins>Evaluating Fit with Adjusted R-squared<ins>
Adjusted R-squared is a crucial statistic in multiple linear regression that modifies the R-squared value to account for the number of predictors in the model.

It provides a more accurate measure of the goodness of fit, especially when comparing models with different numbers of predictors.

### Example Code below:<br>
* Explanation: This code first fits a multiple linear regression model using scikit-learn's `LinearRegression `and then calculates the R-squared using `r2_score`.

 > To compute the adjusted R-squared, we apply the formula that adjusts R-squared based on the number of observations and predictors.

 > This adjusted metric is crucial for evaluating the model, especially when adding more predictors, as it penalizes model complexity to prevent overfitting.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

# Sample dataset: House Size, Number of Rooms, and Age vs House Price
X = np.array([[650, 3, 5], [700, 3, 8], [800, 2, 10], [1200, 4, 2], [950, 3, 5]])  # [Size, Rooms, Age]
y = np.array([300000, 340000, 320000, 500000, 400000])  # Price

# Creating and fitting the multiple linear regression model
model = LinearRegression().fit(X, y)

# Calculating R-squared
y_pred = model.predict(X)
r_squared = r2_score(y, y_pred)

# Calculating Adjusted R-squared
n = X.shape[0]  # number of observations
p = X.shape[1]  # number of predictors
adj_r_squared = 1 - (1 - r_squared) * ((n - 1) / (n - p - 1))

print(f"R-squared: {r_squared}")
print(f"Adjusted R-squared: {adj_r_squared}")
```

## <ins>Feature Selection Through Standardization<ins>
Standardization transforms data to have a mean of zero and a standard deviation of one, ensuring that features with large scales don’t dominate those with small scales in models that are sensitive to the magnitude of variables. Post-standardization, feature selection becomes more straightforward, allowing for an unbiased evaluation of each feature's contribution to the model.

### Example Code below:<br>
* Explanation: This script first standardizes the entire dataset to ensure each feature contributes equally to the analysis.
> It then splits the data into training and test sets to maintain evaluation integrity.

> Using `SelectKBest.` with `f_regression` , the script selects the top 5 features that have the most significant relationship with the target variable based on the training set.<br>
<br>
> The `get_support` method identifies the indices of these top features, allowing us to understand which features contribute most significantly to predicting the target.

```python
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split

# Loading the diabetes dataset
X, y = load_diabetes(return_X_y=True)

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting dataset to avoid leakage from the test set during feature selection
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Feature Selection using SelectKBest
selector = SelectKBest(score_func=f_regression, k=5)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Identifying selected features
selected_features = selector.get_support(indices=True)

print(f"Selected features indices: {selected_features}")
```

## <ins>Logistic vs. Logit Function<ins>
The logistic function is an S-shaped curve that can take any real-valued number and map it between 0 and 1, making it useful for a model predicting probabilities. The logit function is the inverse of the logistic function, used to express the log odds of the probability.

### Example Code below:<br>
* Explanation: This example illustrates the logistic and logit functions conceptually using matplotlib for visualization.

```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic Function
def logistic_function(x):
    return 1 / (1 + np.exp(-x))

# Logit Function
def logit_function(p):
    return np.log(p / (1 - p))

# Generating values
x_values = np.linspace(-10, 10, 100)
logistic_values = logistic_function(x_values)
logit_values = logit_function(logistic_values[1:-1])  # Avoid division by zero

# Plotting
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x_values, logistic_values, label='Logistic Function')
plt.title('Logistic Function')
plt.subplot(1, 2, 2)
plt.plot(logistic_values[1:-1], logit_values, label='Logit Function', color='red')
plt.title('Logit Function')
plt.show()
```

## <ins>Selecting the Right Number of Clusters: How to Choose the Number of Clusters<ins>
Selecting the right number of clusters impacts the interpretability and effectiveness of clustering results. Various methods exist for determining the optimal number of clusters, such as the elbow method and silhouette score.

### Example Code below:<br>
* Explanation: This code calculates the silhouette score for different numbers of clusters and plots the results to help visualize the optimal number of clusters.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Define range of clusters
range_n_clusters = [2, 3, 4, 5]
silhouette_scores = []

# Calculate silhouette scores for different numbers of clusters
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, The average silhouette_score is : {silhouette_avg}")

# Plot silhouette scores
plt.plot(range_n_clusters, silhouette_scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Numbers of Clusters')
plt.show()
```

## <ins>The Dilemma: To Standardize or Not to Standardize<ins>
Standardization helps to ensure that all features contribute equally to the clustering process, especially when the features have different scales or units.

### Example Code below:<br>
* Explanation: This code compares K-means clustering results on the same dataset with and without standardization. It illustrates how standardizing features can affect the clustering outcome.

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data with features on different scales
X = np.array([[1, 2], [1, 4], [1, 0], [4, 2], [4, 4], [4, 0]])

# Apply K-means clustering without standardization
kmeans_no_std = KMeans(n_clusters=2)
kmeans_no_std.fit(X)

# Apply K-means clustering with standardization
X_std = (X - X.mean(axis=0)) / X.std(axis=0)
kmeans_std = KMeans(n_clusters=2)
kmeans_std.fit(X_std)

# Visualize clustering results
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_no_std.labels_, cmap='viridis')
plt.title('Clustering without Standardization')

plt.subplot(1, 2, 2)
plt.scatter(X_std[:, 0], X_std[:, 1], c=kmeans_std.labels_, cmap='viridis')
plt.title('Clustering with Standardization')

plt.show()
```

## <ins>Unveiling the Interplay between Clustering and Regression<ins>
Understanding the relationship between clustering and regression enables the development of hybrid models that leverage both techniques to improve predictive accuracy and interpretability.

### Example Code below:<br>
* Explanation: This code illustrates how to integrate clustering (K-means) and regression to build a predictive model.
> It trains separate linear regression models within each cluster identified by K-means and predicts the target variable for new data points based on the cluster they belong to.

```python
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
import numpy as np

# Generate synthetic data for clustering and regression
X = np.array([[1, 2], [1, 4], [1, 0], [4, 2], [4, 4], [4, 0]])
y = np.array([3, 5, 1, 6, 8, 2])

# Apply K-means clustering
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

# Apply linear regression within each cluster
regressors = []
for cluster in range(kmeans.n_clusters):
    mask = kmeans.labels_ == cluster
    X_cluster = X[mask]
    y_cluster = y[mask]
    regressor = LinearRegression().fit(X_cluster, y_cluster)
    regressors.append(regressor)

# Predict using the appropriate regressor for each data point
y_pred = np.zeros_like(y)
for cluster, regressor in enumerate(regressors):
    mask = kmeans.labels_ == cluster
    X_cluster = X[mask]
    y_pred[mask] = regressor.predict(X_cluster)

# Evaluate the performance of the combined model
mse = np.mean((y - y_pred) ** 2)
print("Mean Squared Error:", mse)
```
